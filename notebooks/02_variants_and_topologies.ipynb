# 2. Variants and topologies

turboswarm ships three velocity variants (`inertia`, `constriction`, `fips`) and four topologies (`global`, `ring`, `vonneumann`, `random`). This notebook compares them on the multimodal Rastrigin function.

```python
import matplotlib.pyplot as plt
import turboswarm as pso

bound, _ = pso.benchmark_info("rastrigin")
bounds = [(-bound, bound)] * 2
```

## Comparing velocity variants

FIPS performs best with a local topology, so it is paired with the ring.

```python
runs = {}
for velocity in ["inertia", "constriction", "fips"]:
    topology = "ring" if velocity == "fips" else "global"
    runs[f"{velocity}/{topology}"] = pso.minimize(
        "rastrigin", bounds=bounds, velocity=velocity, topology=topology,
        n_particles=40, max_iter=300, seed=7,
    )
for name, r in runs.items():
    print(f"{name:22s} best = {r.best_value:.3e}")
```

```python
pso.viz.compare(runs)
plt.show()
```

## Comparing topologies

Same variant (inertia) and seed, different social structure.

```python
topo_runs = {
    t: pso.minimize("ackley", bounds=[(-32.768, 32.768)] * 2, topology=t,
                    n_particles=49, max_iter=300, seed=3)
    for t in ["global", "ring", "vonneumann", "random"]
}
for name, r in topo_runs.items():
    print(f"{name:12s} best = {r.best_value:.3e}")
pso.viz.compare(topo_runs)
plt.show()
```
