# 4. Multi-objective optimization (MOPSO)

With conflicting objectives there is no single best, but a Pareto front of trade-offs. `minimize_multi` returns an approximation of it.

```python
import matplotlib.pyplot as plt
import turboswarm as pso
```

## A bi-objective problem

`f1` pulls toward 0, `f2` toward 2 — the Pareto-optimal set is `x in [0, 2]`.

```python
front = pso.minimize_multi(
    lambda x: [sum(xi ** 2 for xi in x), sum((xi - 2) ** 2 for xi in x)],
    bounds=[(-5, 5)] * 2,
    n_particles=100, max_iter=100, archive_size=100, seed=42,
)
print("front size:", len(front))
```

## Plotting the front

The objective space shows the characteristic trade-off curve.

```python
pso.viz.plot_pareto(front)
plt.show()
```

## Inspecting solutions

Each solution carries its decision vector and its objective values.

```python
for pos, obj in list(zip(front.positions, front.objectives))[:5]:
    print("x =", [round(v, 2) for v in pos], "-> f =", [round(o, 2) for o in obj])
```
