# 1. Quickstart

The basics of `turboswarm`: minimize a benchmark, optimize your own function, and read the result. Run `maturin develop --release` first.

```python
import matplotlib.pyplot as plt
import turboswarm as pso
```

## A native benchmark

Passing the name of a built-in benchmark runs the objective in Rust (no GIL).

```python
r = pso.minimize("rastrigin", bounds=[(-5.12, 5.12)] * 2, seed=42)
print("best value   :", r.best_value)
print("best position:", r.best_position)
print("evaluations  :", r.evaluations)
print("stop reason  :", r.stop_reason)
```

## Your own objective

Any callable `f(list) -> float` works. The optimum of this shifted sphere is at (2, 2, 2).

```python
r = pso.minimize(lambda x: sum((xi - 2) ** 2 for xi in x), bounds=[(-10, 10)] * 3, seed=1)
print([round(v, 4) for v in r.best_position], "->", r.best_value)
```

## The convergence curve

`record_history=True` (the default) keeps the best value per iteration.

```python
r = pso.minimize("rastrigin", bounds=[(-5.12, 5.12)] * 2, seed=42)
pso.viz.plot_convergence(r)
plt.show()
```
