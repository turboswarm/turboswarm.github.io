# 3. Integer, mixed and constrained problems

turboswarm handles real, integer, binary and mixed variables, plus inequality constraints via a penalty.

```python
import turboswarm as pso
```

## Integer variables

`integer=True` discretizes at evaluation time. Optimum at the integer point (3, -2).

```python
r = pso.minimize(lambda x: (x[0] - 3) ** 2 + (x[1] + 2) ** 2,
                 bounds=[(-10, 10)] * 2, integer=True, seed=5)
print("integer argmin:", r.best_position, "value:", r.best_value)
```

## A binary 0/1 knapsack with a constraint

`binary=True` for {0,1} variables; `constraints` adds a quadratic penalty for any `g(x) > 0`.

```python
values, weights, cap = [10, 13, 18, 31, 7, 15], [2, 3, 4, 7, 1, 3], 10
r = pso.minimize(
    lambda x: -sum(v * xi for v, xi in zip(values, x)),
    bounds=[(0, 1)] * len(values), binary=True,
    constraints=[lambda x: sum(w * xi for w, xi in zip(weights, x)) - cap],
    n_particles=50, max_iter=200, seed=1,
)
chosen = [int(b) for b in r.best_position]
print("chosen:", chosen)
print("value :", sum(v for v, c in zip(values, chosen) if c))
print("weight:", sum(w for w, c in zip(weights, chosen) if c), "(cap", cap, ")")
```

## Mixed variable types

One dimension each: real, integer, binary. Optimum (1.5, 3, 1).

```python
r = pso.minimize(
    lambda x: (x[0] - 1.5) ** 2 + (x[1] - 3) ** 2 + (x[2] - 1) ** 2,
    bounds=[(-5, 5), (-10, 10), (0, 1)],
    var_types=["real", "integer", "binary"], seed=7,
)
print("mixed argmin:", [round(v, 3) for v in r.best_position])
```
